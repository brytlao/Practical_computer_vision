**Smoothing**

```
# import dependencies
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import cv2
```

```
# load image
img_name = 'dragon.png'
img = cv2.imread(img_name) # change to your own uploaded image

# convert image to grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# resize image
gray = cv2.resize(gray,(160,120)) # ratio 4:3

# apply smoothing operation
cv2.blur(gray,(3,3))

# create grid for plotting
xx, yy = np.mgrid[0:gray.shape[0],0:gray.shape[1]]

# create figure
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(xx,yy,gray,rstride=1,cstride=1,cmap=plt.cm.gray,linewidth=1)
plt.show()
```

# **Image color conversions**

**Reading an image**

```
# import dependencies
import cv2
from google.colab.patches import cv2_imshow # alternative to cv2.imshow in colaboratory
```

```
# load image
img_name = 'dragon.png'
img = cv2.imread(img_name)

# display previous image
cv2_imshow(img)
```

**Grayscale**

```
# convert image to grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

**HSV and HLS**

```
# convert image to hsv
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)

# convert image to hls
hls = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
cv2_imshow(hls)
```

**LAB color space**

```
# convert image to lab
lab = cv2.cvtColor(img,cv2.COLOR_BGR2LAB)
cv2_imshow(lab)
```
